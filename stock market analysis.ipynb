```python
#when using a kernel density estimator to smooth the histogram
```

```python
tesla['returns'].plot(kind = 'kde', label = 'Tesla', figsize = (15,6))
ford['returns'].plot(kind = 'kde', label = 'Ford')
gm['returns'].plot(kind = 'kde', label = 'GM')
plt.legend()
plt.show()
```

```python
ford['Cumulative_Returns'] = (1 + ford['returns']).cumprod()
gm['Cumulative_Returns'] = (1 + gm['returns']).cumprod()
tesla['Cumulative_Returns'] = (1 + tesla['returns']).cumprod()
```

```python
ford.head()
```

```python
ford['Cumulative_Returns'].plot(label = 'Ford', figsize = (15,8))
tesla['Cumulative_Returns'].plot(label = 'Tesla')
gm['Cumulative_Returns'].plot(label = 'gm')
plt.legend()
plt.title('Cumulative Returns')
plt.show()
```
